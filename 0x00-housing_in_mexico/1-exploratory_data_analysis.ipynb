### Exploratory Data Analysis (EDA)


- Exploratory data analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
```

#### Import data

```python
df = pd.read_csv("data/mexico-real-estate-clean.csv")
df.shape
df.head()
```

#### Location data: Scatter Mapbox Plot

```python
""" 
Using plotly to create the figure from df data.
Map will be centered on Mexico City.
The map displays price when hovering mouse over house
"""
fig=px.scatter_mapbox(
    df, 
    lat="lat",
    lon="lon",
    center={"lat": 19.43, "lon": -99.13},  
    width=600, 
    height=600,  
    hover_data=["price_usd"],  
)

fig.update_layout(mapbox_style="open-street-map")


fig.show()
```

Histogram of area

```python
plt.hist(df["area_m2"]);
plt.xlabel("Area [sq meters]")
plt.ylabel("Frequency")
plt.title("Distribution of Home Sizes")
```

Box plot of Area

```python
plt.boxplot(df["area_m2"]);
plt.xlabel("Area [sq meters]")
plt.title("Distribution of Home Sizes")
```

Histogram of Price

```python
plt.hist(df["price_usd"]);
plt.xlabel("Price [USD]")
plt.ylabel("Frequency")
plt.title("Distribution of Home Prices")
```

Boxplot of price

```python
plt.boxplot(df["price_usd"]);
plt.xlabel("Price [USD]")
plt.title("Distribution of Home Prices")
```
