**Overall Goal**: To predict whether or not a company will go bankrupt based on several financial indicators.

- This project focuses on acquaring data and have it in a usable format.

```python
import gzip
import json

import pandas as pd
from ucimlrepo import fetch_ucirepo 
```

### Prepare Data

Option 1: Assuming the data was stored in a gzipped file:

```python
def wrangle(filename):
    # Open compressed file and load into a dict
    with gzip.open(filename) as f:
        data= json.load(f)
        
    # convert the data into a pandas DataFrame
    df = pd.DataFrame().from_dict(data["data"]).set_index("company_id")
    
    return df
```
